```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
customers = pd.read_csv('Customers.csv')
products = pd.read_csv('Products.csv')
```

```python
customers.head()
```

```python
products.head()
```

```python
transactions = pd.read_csv('Transactions.csv')
transactions.head()
```

```python
df = pd.merge(transactions, customers, on='CustomerID').merge(products, on='ProductID')
df.head()
```

```python
df = df.drop(['TransactionID','TransactionDate','Quantity','Price_x','Price_y','CustomerName','SignupDate','ProductName'],axis=1)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le1 = LabelEncoder()
df['Region'] = le.fit_transform(df['Region'])
df['Category'] = le.fit_transform(df['Category'])
df['ProductID'] = le.fit_transform(df['ProductID'])
df['CustomerID'] = le1.fit_transform(df['CustomerID'])
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['TotalValue'] = scaler.fit_transform(df[['TotalValue']])
df.head()
```

```python
df1 = df[['TotalValue','Region','Category']]
df1.head()
```

```python
df2 = df.pivot_table(index='CustomerID',columns='ProductID',values='TotalValue')
df2.head()
```

```python
df2.fillna(0,inplace=True)
df2.head()
```

```python
df2.idxmax(axis=1)[:10]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(df1,df1)
cosine_sim
```

```python
def look_a_like(anime_title, top_n=3, threshold=0.8):
    anime_index = df[df['CustomerID'] == anime_title].index[0]
    sim_scores = list(enumerate(cosine_sim[anime_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = [x for x in sim_scores if x[1] > threshold]
    sim_scores = sim_scores[1:top_n+1]
    anime_indices = [i[0] for i in sim_scores]
    similar_score = [i[1] for i in sim_scores]
    customer_ids = df.iloc[anime_indices]['CustomerID'].tolist()

    recommendations = list(customer_ids)+list(similar_score)

    return recommendations
```

```python
look_a_like(0, top_n=3, threshold=0.8)
```

```python
customer_ids = set(df['CustomerID'])
customer_ids = list(customer_ids)[:20]
customer_ids
```

```python
lookalike_map = {}
for cust_id in customer_ids:
    lookalike_map[cust_id] = look_a_like(cust_id, top_n=3)
lookalike_map
```

```python
lookalike_df = pd.DataFrame.from_dict(lookalike_map, orient='index')
lookalike_df.columns = ['Lookalike 1', 'Lookalike 2', 'Lookalike 3','Lookalike 1 Score','Lookalike 2 Score','Lookalike 3 Score']
```

```python
lookalike_df
```

```python
lookalike_df.index
```

```python
lookalike_df.index = le1.inverse_transform(lookalike_df.index)
lookalike_df['Lookalike 1'] = le1.inverse_transform(lookalike_df['Lookalike 1'])
lookalike_df['Lookalike 2'] = le1.inverse_transform(lookalike_df['Lookalike 2'])
lookalike_df['Lookalike 3'] = le1.inverse_transform(lookalike_df['Lookalike 3'])
lookalike_df
```

```python
lookalike_df.to_csv('Lookalike.csv')
```
